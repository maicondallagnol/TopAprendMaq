# 1. Dados

O conjunto de dados contém 9568 pontos de dados coletados de uma Central Elétrica de Ciclo Combinado ao longo de 6 anos (2006-2011), quando a usina foi configurada para funcionar com carga total. Os recursos consistem em variáveis ambientais médias por hora Temperatura (T), Pressão ambiente (AP), Umidade Relativa (UR) e Exaustão de Vácuo (V) para prever a produção líquida de energia elétrica (EP) da planta. 
Uma usina de energia de ciclo combinado (CCPP) é composta de turbinas a gás (GT), turbinas a vapor (ST) e geradores de vapor de recuperação de calor. Em um CCPP, a eletricidade é gerada por turbinas a gás e a vapor, que são combinadas em um ciclo e transferidas de uma turbina para outra. Enquanto o vácuo é coletado e tem efeito sobre a turbina a vapor, outras três variáveis ​​ambientais afetam o desempenho do GT.

Fonte: https://archive.ics.uci.edu/ml/datasets/Combined+Cycle+Power+Plant

## 1.1 Informações sobre os dados:

#### Atributos:
- Temperatura (T) na faixa de 1,81 ° C e 37,11 ° C, 
- Pressão ambiente (AP) na faixa de 992,89-1033,30 milibar, 
- Umidade Relativa (UR) na faixa de 25,56% a 100,16 % 
- Vácuo de exaustão (V) na faixa de 25,36-81,56 cm Hg 
- Saída de energia elétrica horária líquida (EP) 420,26-495,76 MW

## Escalonando

```python
scaler = StandardScaler().fit(data_raw)
data_scaled = scaler.transform(data_raw)
```

```python
data_scaled_df = pd.DataFrame(data_scaled, columns=data_raw.columns)
```

```python
data_scaled_df.head()
```

```python
data_scaled_df.plot.box()
```

## Utilidades

```python
lista_metricas_treino = []
lista_metricas_teste = []
```

```python
def metricas(y_true, y_pred, alg):
    r2 = r2_score(y_true, y_pred)
    eqm = mean_squared_error(y_true, y_pred)
    seq = len(y_true)*eqm
    reqm = math.sqrt(eqm)
    
    return {'Algoritmo':alg, 'R2':r2, 'EQM':eqm, 'REQM':reqm, 'SEQ':seq}
```

## Separando conjuntos de Treino e Teste

Para a separação utilizou-se do train_test_split que divide o conjunto em treino e teste aleatóriamente

```python
train, test = train_test_split(data_scaled_df, test_size = 0.2, shuffle=True)

x_train = train.drop(columns=['Temperature (C)'])
y_train = train['Temperature (C)']

x_test = test.drop(columns=['Temperature (C)'])
y_test = test['Temperature (C)']
```

## Aplicando a Regressão

### Regressão Linear

```python
lire = LinearRegression()
```

```python
lire.fit(x_train, y_train)
```

## Avaliação para Teste

```python
y_pred = lire.predict(x_test)
linear_metricas = metricas(y_test, y_pred, 'Regressão Linear - Teste')
lista_metricas_teste.append(linear_metricas)
```

```python
plt.scatter(x_test['Humidity'], y_test,  color='black')
plt.scatter(x_test['Humidity'], y_pred, color='yellow')
plt.xlabel("Humidity")
plt.ylabel("Temperature")
plt.title('Predito e Original',fontsize=15)
plt.legend(['Original', 'Predito'])
plt.show()
```

## Avaliação para Treino

```python
y_pred = lire.predict(x_train)
linear_metricas = metricas(y_train, y_pred, 'Regressão Linear - Treino')
lista_metricas_treino.append(linear_metricas)
```

```python
plt.scatter(x_train['Humidity'], y_train,  color='black')
plt.scatter(x_train['Humidity'], y_pred, color='yellow')
plt.xlabel("Humidity")
plt.ylabel("Temperature")
plt.title('Predito e Original',fontsize=15)
plt.legend(['Original', 'Predito'])
plt.show()
```
